```python
from transformers import pipeline, set_seed
# Here, 'text-generation' tells the pipeline to create a text generation model, and model='gpt2'
# specifies that you want to use the GPT-2 model for this task.

# Unlike the above where there was no. model supplied we are specifying the model as model='gpt2'
generator = pipeline('text-generation', model='gpt2')
```

```python
set_seed(100)
input_text = "This project is for computional social science, a course at UC Berkeley, where we are estimating the difference between Democrat and Republican legislators tweets score for"
# specifying the maximum length of the output, and the number of different sequences you want. AKA this is likely texts we would see
output = generator(input_text, max_length=50, num_return_sequences=5)
for result in output:
    print(result['generated_text'])
    print("----")
```
