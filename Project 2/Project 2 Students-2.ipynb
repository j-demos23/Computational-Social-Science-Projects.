#### d. Check for duplicate columns

There are a lot of columns in this data frame. Let's see if there are any are duplicates. Note that Pandas will not allow them to have the same exact column name, so they will likely be distinct on column name but will be copies otherwise.

```python
#
# Check for duplicate columns
#-----------
# I used Google to figure this out, and adapted this example for our purposes:  
# source: https://thispointer.com/how-to-find-drop-duplicate-columns-in-a-dataframe-python-pandas/ 
def getDuplicateColumns(df):
    '''
    Get a list of duplicate columns.
    It will iterate over all the columns in dataframe and find the columns whose contents are duplicate.
    :param df: Dataframe object
    :return: List of columns whose contents are duplicates.
    '''
    duplicateColumnNames = set()
    # Iterate over all the columns in dataframe
    for x in range(df.shape[1]):
        # Select column at xth index.
        col = df.iloc[:, x]
        # Iterate over all the columns in DataFrame from (x+1)th index till end
        for y in range(x + 1, df.shape[1]):
            # Select column at yth index.
            otherCol = df.iloc[:, y]
            # Check if two columns at x 7 y index are equal
            if col.equals(otherCol):
                duplicateColumnNames.add(df.columns.values[y])
    return list(duplicateColumnNames)

duplicateColumnNames = list(getDuplicateColumns(diabetes))
print('Duplicate Columns are as follows: ')
duplicateColumnNames
```

```python
#
# drop columns from duplicates list
#-----------
# now drop list of duplicate features from our df using the .drop() method
diabetes = diabetes.drop(columns=duplicateColumnNames) 
```

By minimizing the MSE, linear regression finds the line or hyperplane that best describes the data, and this often results in better predictions than guessing a constant value. The minimizing process ensures that the line provides the best compromise in terms of minimizing prediction errors. However OLS models do not deal with multicollinearity which ma arise in a dataset such as this given the high amount of predicctors included in the model that are similar in nature (i.e various predictors for race, sex etc). 

Ridge and Lasso

The lamda hyper parameter helps us deal with caes of bias and vairance in different ways. Both models utilize a lamda tuning parameter.  The ridge regresion has uses l2 norm or the summation of square betas in a model which primarly helps us reduce the vairance of our model. Ridge does this my shrinking our coefficents towards zero values.This is best suited for dealing with multicolinearity However LASSO models use l1 term which is the summation of the absolute values of our beta coefficents. This helps with our vairance but also  providse the benifet of feature selction as some of our coefficents will actually be reduced to zero and not just towards zero. Therefore LASSO is better than Ridge in terms of regulation overfitting our model. These beniefets are useful for our data and question becuase there are various sets of predictors that are related to one another than could lead to multicolliniarity and the sheer amount of predictors can easily lead to over fitting the model. ### 5.2 Impement a cross-validation approach

Finally, implement a cross-validation approach for your best model and evaluate the results.

```python
#
# Run cross-validation
#-----------
#
# RIDGE 
#-----------
# We will run 9*2*4 = 72 tests, each with 3-fold cross validation

# specify the hyperparameters
param_grid = {'alpha': np.arange(.1, 1, .1),
              'fit_intercept': [True, False],
              'solver': ['auto', 'svd', 'cholesky', 'lsqr']}

# execute the grid search
ridge_grid_reg = GridSearchCV(ridge_reg,  # model to be tuned
                              param_grid, # parameters to be searched as specified above
                              cv=3)       # 3-fold cross-validation to be used during hypertuning

# fit the tuning on the training data
ridge_grid_reg.fit(X_train, y_train)

# select the best performing model and predict with that on validation dataset 
best_index = np.argmax(ridge_grid_reg.cv_results_["mean_test_score"])  # find the best performing model
best_ridge_pred = ridge_grid_reg.best_estimator_.predict(X_validate)   # find best estimator and predict on validate

# print the results  
print(ridge_grid_reg.cv_results_["params"][best_index])
print('Best CV R^2:', max(ridge_grid_reg.cv_results_["mean_test_score"]))
print('Validation R^2:', ridge_grid_reg.score(X_validate, y_validate))
print('Validation RMSE', rmse(best_ridge_pred, y_validate))
```

1. What is bias-variance tradeoff? Why is it relevant to machine learning problems like this one?



The bias-variance tradeoff in machine learning that involves balancing two key sources of error in predictive modeling. Bias refers to the error introduced by overly simplistic assumptions (limited predictors) in the learning algorithm. Models with high bias are typically too simple to capture complex patterns in the data and may underfit the training data, leading to poor performance. On the other hand, variance relates to the error introduced by overly complex models that are highly sensitive to variations in the training data. Such models can overfit the training data, capturing noise and failing to generalize well to new, unseen data.

Finding the right balance between bias and variance is essential for building effective machine learning models. This balance can be achieved through careful model selection, hyperparameter tuning, and the use of techniques like cross-validation. By understanding and managing the bias-variance tradeoff, researcehrs can create models that generalize well, make accurate predictions on new data, and avoid underfitting and overfitting.
